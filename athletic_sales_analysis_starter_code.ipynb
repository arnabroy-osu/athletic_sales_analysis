```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the 2020 CSV files into DataFrame
df_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')

# Read the 2021 CSV files into DataFrame
df_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')


```

```python
# Display the 2020 sales DataFrame
df_2020.head()
```

```python
# Display the 2021 sales DataFrame
df_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes

```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

# Arnab's Comments
# I used concat function for default Axis=0 to combine the two DataFrames on the rows 
# Reason: All the columns have same data types and column names.
# I used the ignore_index=True to clear the source index and reset it to the new index.
# If you don't use ignore_index=True, the combined DataFrame will have duplicate index values.


df_combined = pd.concat([df_2020, df_2021], ignore_index=True) 
df_combined.head()



```

```python
# Check if any values are null in df_combined DataFrame.

# Arnab's Comments
# I used isnull().values.any() to check if any values are null in df_combined DataFrame.
# If this returns True, we can drop NULLs using dropna() function.
df_combined.isnull().values.any()



```

```python
# Check the data type of each column
df_combined.dtypes

```

```python
# Convert the "invoice_date" to a datetime datatype

# Arnab's Comments
# invoice_date field has Date Only values.
df_combined["invoice_date"] = pd.to_datetime(df_combined["invoice_date"])
df_combined.head()


```

```python
# Confirm that the "invoice_date" data type has been changed.
df_combined.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

# Arnab's Comments
# Using groupby function to calculate sum(units_sold) by region, state, and city 
# and renamed the sum to "Total_Products_Sold" in the same step.
# Sorted the results by Total_Products_Sold in descending order.
df_Total_Products_Sold_Groupby = df_combined.groupby(["region", "state", "city"]).agg({"units_sold": "sum"}).rename(columns={"units_sold": "Total_Products_Sold"})
df_Total_Products_Sold_Groupby = df_Total_Products_Sold_Groupby.sort_values(by="Total_Products_Sold", ascending=False)



# Show the top 5 results.
df_Total_Products_Sold_Groupby.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold"

# Arnab's Comments
# Using Pandas pivot_table to calculate sum(units_sold) for region, state, and city
# and renamed the units_sold to "Total_Products_Sold" in the same step.
df_Total_Products_Sold_Pivot = pd.pivot_table(df_combined, 
                                        index=["region", "state", "city"], 
                                        values=["units_sold"], 
                                        aggfunc="sum"
                                        ).rename(columns={"units_sold": "Total_Products_Sold"})
df_Total_Products_Sold_Pivot.head()

# Sort the results by Total_Products_Sold in descending order
df_Total_Products_Sold_Pivot = df_Total_Products_Sold_Pivot.sort_values(by="Total_Products_Sold", ascending=False)


# Show the top 5 results.
df_Total_Products_Sold_Pivot.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

# Arnab's Comments
# Using groupby function to calculate sum(total_sales) by region, state, and city
# and renamed the total_sales to "Total Sales" in the same step.
# Sorted the results by Total Sales in descending order.
df_Total_Sales_Groupby = df_combined.groupby(["region", "state", "city"]).agg({"total_sales": "sum"}).rename(columns={"total_sales": "Total Sales"})
df_Total_Sales_Groupby = df_Total_Sales_Groupby.sort_values(by="Total Sales", ascending=False)

# Show the top 5 results.
df_Total_Sales_Groupby.head(5)

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"

# Arnab's Comments
# Using Pandas pivot_table to calculate sum(total_sales) for region, state, and city
# and renamed the total_sales to "Total Sales" in the same step.
# Sorted the results by Total Sales in descending order.
df_Total_Sales_Pivot = pd.pivot_table(df_combined, 
                                        index=["region", "state", "city"], 
                                        values=["total_sales"], 
                                        aggfunc="sum"
                                        ).rename(columns={"total_sales": "Total Sales"})

df_Total_Sales_Pivot = df_Total_Sales_Pivot.sort_values(by="Total Sales", ascending=False)
                                        

# Show the top 5 results.
df_Total_Sales_Pivot.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
df_Women_Athletic_Footwear = df_combined[df_combined["product"] == "Women's Athletic Footwear"]


# Arnab's Comments
# Used unique function to confirm that we have only one product in df_Women_Athletic_Footwear DataFrame.
df_Women_Athletic_Footwear["product"].unique()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Arnab's Comments
# Using groupby function to calculate sum(units_sold) by retailer, region, state, and city in df_Women_Athletic_Footwear DataFrame
# and renamed the units_sold to "Womens_Footwear_Units_Sold" in the same step.
# Sorted the results by "Womens_Footwear_Units_Sold" in descending order.

df_Total_WAF_Sold_Groupby = df_Women_Athletic_Footwear.groupby(["retailer", "region", "state", "city"]).agg({"units_sold": "sum"}).rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
df_Total_WAF_Sold_Groupby = df_Total_WAF_Sold_Groupby.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)


# Show the top 5 results.
df_Total_WAF_Sold_Groupby.head(5)

```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Arnab's Comments
# Using Pandas pivot_table to calculate sum(units_sold) by retailer, region, state, and city in df_Women_Athletic_Footwear DataFrame
# and renamed the units_sold to "Womens_Footwear_Units_Sold" in the same step.
# Sorted the results by "Womens_Footwear_Units_Sold" in descending order
df_Total_WAF_Sold_Pivot = pd.pivot_table(df_Women_Athletic_Footwear, 
                                        index=["retailer", "region", "state", "city"], 
                                        values=["units_sold"], 
                                        aggfunc="sum"
                                        ).rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})
df_Total_WAF_Sold_Pivot = df_Total_WAF_Sold_Pivot.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)



# Show the top 5 results.
df_Total_WAF_Sold_Pivot.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
df_Women_Athletic_Footwear.head()

# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.

df_WAF_Sales_Pivot = pd.pivot_table(df_Women_Athletic_Footwear, 
                                        index=["invoice_date"], 
                                        values=["total_sales"],
                                        aggfunc="sum"
                                        ).rename(columns={"total_sales": "Total Sales"})
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
df_WAF_Daily_Sales_Pivot = df_WAF_Sales_Pivot.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
df_WAF_Daily_Sales_Pivot = df_WAF_Daily_Sales_Pivot.sort_values(by="Total Sales", ascending=False)

# show the top 10 days that generated the most women's athletic footwear sales
df_WAF_Daily_Sales_Pivot.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
df_WAF_Weekly_Sales_Pivot = df_WAF_Sales_Pivot.resample('W').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
df_WAF_Weekly_Sales_Pivot = df_WAF_Weekly_Sales_Pivot.sort_values(by="Total Sales", ascending=False)

#  show the top 10 weeks that generated the most women's athletic footwear sales
df_WAF_Weekly_Sales_Pivot.head(10)
```
